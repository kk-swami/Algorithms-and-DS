Given a list of integers (can include negative ints, 0 as well,
 return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.
 
So output is a list of non-duplicated lists

Example : Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]

Insight:

The naive approach is O(n3) time complexity as we generate all triplets and find each sum

The natural approach is O(n2) where we use two points, i from 0 to n-1, j from i to n-1 and check if a[i] + a[j] = -a[k]

However , just implementing this naively in python using lists gave time out issues for cases where n is very large

Trick : use sets everywhere, as finding membership of element in set is O(1) unlike a list where it is O(n)

Of course, unclear how complex converting a list to set is but this is one time

```python
def threeSum(nums) :
    ## edge case, if len(nums)<3, there can be no triplets
    
    nums.sort()    
    n = len(nums)
    numset = set(nums)
    

    
    if n<3:
        return []
    else:
        outlist = set()
        elem12 = set()
        for i in range(0, n):
            elem_1 = nums[i]

            for j in range(i+1,n):
                elem_2 = nums[j]
                elem_12_tup = (elem_1, elem_2)
                if elem_12_tup not in elem12:
                    elem12.update([elem_12_tup])
                    elem_3 = -(nums[i]+nums[j])
                    if elem_3 in numset:
                        if elem_3 in set(nums[(j+1):]):
                            temp_tup = (elem_1, elem_2, elem_3)

                            if temp_tup not in outlist:
                                outlist.update([temp_tup])
    
                
    return list(outlist)           
                
            
        
    
```
